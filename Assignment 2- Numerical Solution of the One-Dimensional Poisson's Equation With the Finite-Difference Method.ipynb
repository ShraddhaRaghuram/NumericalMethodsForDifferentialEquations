```python
#3.
# Solution of 1D Poissonâ€™s equation with FDM
# Shraddha Raghuram (c) 2022
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
#4. (a)
n = 5
xgrid = np.linspace(0,3,n+1)
```

```python
#4. (b)
def func1(x):
    f1=3*x-2
    return f1
def func2(x):
    f2=(x**2) +3*x-2
    return f2
```

```python
#4. (c)
f1 = np.reshape(func1(xgrid),(6,1))
f2 =np.reshape( func2(xgrid),(6,1))
print (f1)
```

```python
#4. (d)
plt.figure()
plt.plot(xgrid, f1, 'bo', label='$f_1$', linestyle='-')
plt.plot(xgrid, f2, 'ro', label='$f_2$', linestyle='-')
plt.legend()
plt.title('Right-hand side functions')
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
#4. (g)
def u1(x):
    u1=((-x**3)/2)+(x**2)+(3*x/2)+1
    return u1
def u2(x):
    u2=-((x**4)/12)+((-x**3)/2)+(x**2)+(15*x/4)+1
    return u2
u1ex = u1(xgrid)
u2ex = u2(xgrid)
plt.figure()
plt.plot(xgrid, u1ex, 'bo', label='$f_1$', linestyle='-')
plt.plot(xgrid, u2ex, 'ro', label='$f_2$', linestyle='-')
plt.legend()
plt.title('Exact solutions')
plt.xlabel('x')
plt.ylabel('u(x)')
u1ex=u1ex.reshape(6,1)
u2ex=u2ex.reshape(6,1)
```

```python
#5. (b)
A=np.zeros((n-1,n-1))
mid=np.full((n-1,n-1),-2*(n**2)/9)
adj=np.full((n-2,n-2),1*(n**2)/9)
adjb=A[:n-1,0:n-1]+np.diag(np.diag(adj), k=-1)
adja=A[:n-1,0:n-1]+np.diag(np.diag(adj), k=1)
A=A+np.diag(np.diag(mid))+adja+adjb
print(A)
```

```python
#5. (c)
struc=plt.spy(A, marker='o', color='green')
```

```python
#5. (d)
val,vec=np.linalg.eig(A)
print(val)
```

```python
#6. (a)
a=np.array([0,(n**2)/9,0,0,(n**2)/9,0])
a=a.reshape(6,1)
f1rhs= f1+a
f2rhs=f2+a
```

```python
#6. (b)
u1=np.linalg.solve(A,f1rhs[1:5])
u2=np.linalg.solve(A,f2rhs[1:5])
```
